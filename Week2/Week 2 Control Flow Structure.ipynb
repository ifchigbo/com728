## Selection Control/ Decision Control

### Introduction

- <font size= "3">A control structure that **alters the sequential flow** by branching to specific statements based on a **condition or decision**</font>
- <font size= "3">We use relational operators to set a **condition or decision**</font>
- <font size= "3">Decision control structures help programs seem intelligent because they can make responses that correspond to user input</font>
- <font size= "3">Allow programs to handle data based on differing values</font>

### Simple decision

### *1. if statement*
<br>
<font size ="3">A logic that directs the program flow to one or more statements in a code block when the condition is <span style="color: blue;"><b>True</b></span>, or bypasses the code block if the condition is False</font>

**Syntax:**<br>


**If statement example**

```python
'''Program to calculate personal income tax in UK.
   When you earn money from employment, you may need to pay income tax contributions from your earnings.
   Each person will have a tax-free Personal Allowance which is the amount of income you do not have to pay tax on.
   This program can help you to calculate how much income tax you will need to pay.
'''
#The standard personal allowance in the UK for 2022-2023 is GBP 12,570 per year
personal_allowance = 12570
tax_rate=0.2

#Retieve the income earn from a user
income = float(input("Enter your current income in GBP: "))

#Check wether the income is more than personal allowance
if income > personal_allowance:
    income_tax = income * tax_rate
    print(f"Over the year, you will pay GBP {income_tax} income tax from your annual income of GBP {income}.")

print("Calculation complete!")
```

### **2.if-else statement**

- <font size="3">A decision coded as <b>two possible paths</b></font>
- <font size="3">Specify what statements should be executed when the condition of an if statement is evaluated as <span style="color: blue;"><b>False</b></span></font>
- <font size="3">The <span style="color: blue;"><b>else</b></span> keyword designates the alternative execution path</font>
- <font size="3">We can only have at most one else with each if statement</font>

**Syntax:**<br>


**if-else statement example**

```python
'''Program to calculate personal income tax in UK.
   When you earn money from employment, you may need to pay income tax contributions from your earnings.
   Each person will have a tax-free Personal Allowance which is the amount of income you do not have to pay tax on.
   This program can help you to calculate how much income tax you will need to pay.
'''
#The standard personal allowance in the UK for 2022-2023 is GBP 12,570 per year
personal_allowance = 12570
tax_rate = 0.2

#Retieve the income earn from a user
income = float(input("Enter your current income in GBP: "))

#Check wether the income is more than personal allowance, if not you don't need to pay income tax
if income > personal_allowance:
    income_tax = income * tax_rate
    print(f"Over the year, you will pay GBP {income_tax} income tax from your annual income of GBP {income}.")
else:
    print("You do not need to pay income tax")

print("Calculation complete!")
```

### 3. if-elif-else statement

- <font size="3">A structure used to handle algorithms that require <b>multiple conditions<b></font>
- <font size="3">The computer evaluates statements **only until one of the conditions is true**, then skips any remaining elif statements</font>
- <font size="3">The elif is optional and you can have as many as you like</font>

**Syntax:**<br>


**if-elif-else expample**


```python
'''Program to calculate personal income tax in UK.
   When you earn money from employment, you may need to pay income tax contributions from your earnings.
   Each person will have a tax-free Personal Allowance which is the amount of income you do not have to pay tax on.
   - income above £12,571, you'll pay income tax at 20% - known as the basic rate.
   - income above £50,271, you'll pay at 40% - known as the higher rate.
   - income above £150,000, you'll pay 45% (the additional rate). 
   This program can help you to calculate how much income tax you will need to pay.
'''
#The standard personal allowance in the UK for 2022-2023 is GBP 12,570 per year
personal_allowance = 12570
basic_rate = 0.2
higher_rate = 0.4
addition_rate = 0.45


#Retieve the income earn from a user
income = float(input("Enter your current income in GBP: "))

#Check wether the income is in which thresholds and calculate income tax according to the rate
if income > personal_allowance:
    income_tax = income * basic_rate
    print(f"Over the year, you will pay GBP {income_tax} income tax from your annual income of GBP {income}.")
elif income > 50271:
    income_tax = income * higher_rate
    print(f"Over the year, you will pay GBP {income_tax} income tax from your annual income of GBP {income}.")
elif income > 150000:
    income_tax = income * addition_rate
    print(f"Over the year, you will pay GBP {income_tax} income tax from your annual income of GBP {income}.")
else:
    print("You do not need to pay income tax")

print("Calculation complete!")
```

### **Nested-if statement**

- <font size="3">Decisions inside decisions</font>
- <font size="3">This need arises in situations where the decision to be made depends on the outcome of a previous decision</font>
- <font size="3">To implement, use a second indent level for the inner decision structure</font>

**Syntax:**<br>


**Nested-if example**

```python
'''Program to calculate personal income tax in UK.
   When you earn money from employment, you may need to pay income tax contributions from your earnings.
   Each person will have a tax-free Personal Allowance which is the amount of income you do not have to pay tax on.
   - income between £12,571 and £50,270, you'll pay income tax at 20% - known as the basic rate.
   - income between £50,271 and £150,000, you'll pay at 40% - known as the higher rate.
   - income above £150,000, you'll pay 45% (the additional rate).
   This program can help you to calculate how much income tax you will need to pay.
'''
#The standard personal allowance in the UK for 2022-2023 is GBP 12,570 per year
personal_allowance = 12570
basic_rate = 0.2
higher_rate = 0.4
addition_rate = 0.45


#Retieve the income earn from a user
income = float(input("Enter your current income in GBP: "))

#Check wether the income is more than personal allowance. If yes, check in which rate you should pay and 
#calculate income tax according to the rate

if income > personal_allowance:
    if income < 50270:
        income_tax = income * basic_rate
        print(f"Over the year, you will pay GBP {income_tax} income tax from your annual income of GBP {income}.")
    elif income > 50271:
        if income <= 150000:
            income_tax = income * higher_rate
            print(f"Over the year, you will pay GBP {income_tax} income tax from your annual income of GBP {income}.")
        elif income > 150000:
            income_tax = income * addition_rate
            print(f"Over the year, you will pay GBP {income_tax} income tax from your annual income of GBP {income}.")
else:
    print("You do not need to pay income tax")

print("Calculation complete!")
```

### **1. and operator**

- <font size="3">The and operator requires that <b>both</b> conditions be <b>true</b> for the expression to evaluate to <b>True</b></font>
- <font size="3">Truth table: a table summarising the possible outcomes for logical operations</font>

**and operator example**

```python
'''Program to calculate personal income tax in UK.
   When you earn money from employment, you may need to pay income tax contributions from your earnings.
   Each person will have a tax-free Personal Allowance which is the amount of income you do not have to pay tax on.
   - income between £12,571 and £50,270, you'll pay income tax at 20% - known as the basic rate.
   - income between £50,271 and £150,000, you'll pay at 40% - known as the higher rate.
   - income above £150,000, you'll pay 45% (the additional rate). 
   This program can help you to calculate how much income tax you will need to pay.
'''
#The standard personal allowance in the UK for 2022-2023 is GBP 12,570 per year
personal_allowance = 12570
basic_rate = 0.2
higher_rate = 0.4
addition_rate = 0.45


#Retieve the income earn from a user
income = float(input("Enter your current income in GBP: "))

#Check wether the income is in which thresholds and calculate income tax according to the rate
if (income > personal_allowance) and (income < 50270):
    income_tax = income * basic_rate
    print(f"Over the year, you will pay GBP {income_tax} income tax from your annual income of GBP {income}.")
elif (income > 50271) and (income <= 150000):
    income_tax = income * higher_rate
    print(f"Over the year, you will pay GBP {income_tax} income tax from your annual income of GBP {income}.")
elif income > 150000:
    income_tax = income * addition_rate
    print(f"Over the year, you will pay GBP {income_tax} income tax from your annual income of GBP {income}.")
else:
    print("You do not need to pay income tax")

print("Calculation complete!")
```

## Repetition control - loops

- <font size="3">You can visualise a loop as a block of code with an entry point, a few statements that <b>repeat</b>, and an exit point</font>
- <font size="3">Two basic types of repetition control structures:</font>
    - <font size="2">Condition-controlled loops: while-loop</font>
    - <font size="2">Count-controlled loops: for-loop</font>

### 1. Condition-controlled loops: while-loop

- <font size="3">A loop regulated by <b>conditional statements</b> containing <b>Boolean expressions</b> similar to those that control decision structures</font>
- <font size="3">A loop begins with a condition (pre-test)</font>
    - <font size="3">Loop executes only if the test condition is <span style="color: blue;"><b>True</b></span></font>
    - <font size="3">Continues executing until the condition is <span style="color: blue;"><b>False</b></span> or the loop encounters a <span style="color: blue;"><b>break</b></span> statement</font>

**Syntax:**<br>


**while-loop example**

```python
#Weightlifting exercise checking and monitoring program
count = 1
print("Enter the number of reps for this lifting exercise: ")
reps = int(input())
while reps < 1 or reps > 25:
    reps = int(input("Try again. Reps have to be in the range 1 to 25: "))
print(f"OK, you want to do: {reps} reps")
while count <=reps:
    print(f"lift the weight: {count}")
    count = count+1
```

### 2. Count-controlled loops: for-loop

- <font size="3">Repeats a specified number of times</font>
- <font size="3">A count-controlled loop begins with a control statement containing the following parameters:</font>
    - <font size="2">A variable for the <b>loop counter</b> that tracks the number of times a loop has repeated</font>
    - <font size="2">An <b>iterative sequence</b> for the loop counter to use to control the loop.  An iterative squence is data in Python that can be linearly traced e.g., 1,2,3,4</font>
- <font size="3">In Python, we use <span style="color:blue;"><b>range function</b></span> for creating a numerical iterative sequence. It will create a numberical iterative sequence based on its own parameters:</font>
    - <font size="2">A staring number (optional - if the starting point is omitted, Python assumes it is 0)</font>
    - <font size="2">An ending number</font>
    - <font size="2">An incremental value; use to skip numbers in the sequence (optional - if the incremental value is omitted, Python assume it is 1)</font>
- <font size="3">For-loop is appropriate for when you know <b>how many times</b> you want a loop to repeat (i.e., the number of desired iterations (a synonym for repetition))</font>

**Syntax**<br>

<br> <b>Where i is a loop counter variable</b>

**for-loop example and range() function**

```python
#specify only ending number
for i in range(10):
    print(i)
```

<div class="alert alert-block alert-info">
<b>Tip:</b> The specified ending number is never included in the resulting sequence. 
</div>

```python
#specify starting and ending number
for i in range(1,10):
    print(i)
```

```python
#specify starting number, ending number, and an incremental value
for i in range(1,10,2):
    print(i)
```

**for-loop example: Weightlifting exercise position of bench press guidance**

```python
#Collect input for reps from a user
reps = int(input("How many reps do you want to do the bench press?: "))
for i in range(reps):
    print(f"reps: {i+1}")
    print("Lift, pause, lower")
```

<div class="alert alert-block alert-info">
<b>Tip:</b> Another application of loops is to accumulate a total by repeatedly adding values.
</div>

```python
#Checking the total weight for the bench press exercise you perform

#initial variable for a total weight at 0
total_weight = 0

#Ask a user how many reps to do the bench press
reps = int(input("How many reps do you want to do the bench press?:"))

#Ask how much weights to lift (in lbs)
weight = int(input("What size weights are you using (in lbs)?:"))

#accumulate the weight that you are lifting
for i in range(reps):
    total_weight = total_weight + (weight * 2)
    print(f"Rep {i+1}, {total_weight} lbs so far.")

print(f"Your total lift is: {total_weight} lbs.")
```

### **3. Nested loop**

- <font size="3">Loops within loops</font>
- <font size="3">Nested loops contain multiple control statements and loop counters</font>
- <font size="3">Inner and outer loops</font>
    - <font size="2">In pseudocode for a nested loop, the inner loop is indented from the outer loop</font>
    - <font size="2">Each loop control variable must be unique</font>
    - <font size="2">Programmers typically use **j** for the inner loop variable</font>

**Nested loop example: Weightlifting exercise monitor**

```python
total_weight = 0
weight = int(input("What size weights are you using (lbs)? "))
sets = int(input("How many sets? "))
reps = int(input("How many reps per set? "))
for s in range(sets):
    print(f"Set: {s+1}")
    for r in range(reps):
        total_weight = total_weight+(weight*2)
        print(f"Rep: {r+1}, Progress: {total_weight} lbs.")
print(f"Your total lift is: {total_weight} lbs.")
```
